```python
#Text to analyze.
text = """'Early detection of Alzheimer's disease (AD) during its long prodromal period is of paramount importance clinically. Understanding how language changes during this period may provide a deeper understanding of early progression of AD and strengthen early detection when treatments may most effectively halt disease progression (Golde, DeKosky, & Galasko, 2018; Selkoe, 2021). At the same time, such investigation can inform our understanding of language as a cognitive system and crucially advance our knowledge of fundamental brain-language relations.Study of the neuroscientific foundations of the progression of AD, including its prodromal preclinical periods, has advanced widely (Braak & Braak, 1991, 1996; Braak, Thal, Ghebremedhim, & Del Tredicit, 2011; Dickerson et al., 2008, 2011, 2013; Dickerson & Wolk, 2012; Karaman, Mormino, & Sabuncu, 2022; Putcha, Brickhouse, Wolk, & Dickerson, 2019; Sperling et al., 2011). Behaviorally and clinically, Mild Cognitive Impairment (MCI) is defined as a prodromal stage in which there is evidence of cognitive decline that often continues to advance to dementia. Primary difficulty with memory during this period, referred to as MCI amnestic subtype (aMCI), is most likely to be attributable to Alzheimer's disease (AD) pathology (Petersen, 2003, 2004; Rye, Vik, Kocinski, Lundervold, & Lundervold, 2022; Vos et al., 2013; Wolk, Dickerson, & Alzheimer’s Disease Neuroimaging Initiative, 2011). However, while memory deficits are primary, there are also decades of research revealing that deficits in other cognitive domains, including language, also occur during prodromal and clinical periods of AD (see Sherman, Henderson, Flynn, Gair, & Lust, 2021 for recent review). Within the research on language, there are differing views as to the degree and manner in which linguistic factors, syntactic, semantic, or both (Ahmed, Haigh, de Jager, & Garrard, 2013; Chapin, Clarke, Garrard, & Hinzen, 2022; Joyal, Groleau, Bouchard, Wilson, & Fecteau, 2020; Kemper et al., 1993) contribute to this decline.'"""
```

```python
#Import spacey.

import spacy
```

```python
#load the small language model.
processor = spacy.load("en_core_web_sm")
```

```python
#Process text.
processed_text = processor(text)
processed_text
```

```python
#Obtain the number of words in the text.
words = text.split()
number_of_words = len(words)
print("Number of words:", number_of_words)
```

```python
#Split text into distinct sentences.
n = 0
for sentence in processed_text.sents:
    print (n,sentence)
    n += 1
```

```python
#Obtain additional information (noun vs. verb/etc, dictionary version of the word, and the syntactic dependency).
n = 0
for sentence in processed_text.sents:
    for token in sentence:
        print (n, token, token.pos_, token.lemma_, token.dep_)
        n += 1
```

```python
#Obtain number of unique words by coverting them to lowercase + filtering puncutation/stops within the text.
proper_words = [token.text.lower() for token in processed_text if not token.is_stop and not token.is_punct]
proper_words
```

```python
##Question 1
from collections import Counter

#Count the number of unique words that are a part of proper_words + obtain a list of the 25 most common words.
proper_word_freq = Counter(proper_words)
most_common = proper_word_freq.most_common(25)
print(most_common)
```

```python
##Question 2

#Create dictionary to store the frequency of words.
pos_counters = {}

#Loop through the tokens in processed text.
for token in processed_text:
    #Filter stop/punct tokens.
    if not token.is_stop and not token.is_punct:
        #Check if token has not been recorded in pos_counters.
        if token.pos_ not in pos_counters:
            #Create new entry in pos_counters.
            pos_counters[token.pos_] = Counter()
        pos_counters[token.pos_][token.text.lower()] += 1

#Print most common words.
for pos, counter in pos_counters.items():
    print(f"{pos} most common words: {counter.most_common(5)}")
```

```python
##Question 3

#Dependency tree of the word.
def pr_tree(word, level):
    #Skip punc tokens.
    if word.is_punct:
        return
    #obtain syntactic association for all the child tokens to the left of the word.
    for child in word.lefts:
        pr_tree(child, level + 1)
    #Print token + dependency relation.  
    print('\t' * level + word.text + ' - ' + word.dep_)
    #obtain syntactic assocation for all the child tokens to the right of the word.
    for child in word.rights:
        pr_tree(child, level + 1)

#Loop dependency tree through the text to obtain structure of each sentence as it relates to syntactic relations.
for sentence in processed_text.sents:
    pr_tree(sentence.root, 0)
    print('-------------------------------------')
```

```python
##Question 4

#Obtain a list of all the entities in processed_text and print them + their types.
for entity in processed_text.ents:
    print(entity, entity.label_)
```

```python
##Question 5

#Obtain a list of all the entities in processed_text + print entitiy & dependency relation.
for ent in processed_text.ents:
    print(f"entity: {ent.text}, dependency: {ent.root.dep_}")
```

```python
##Question 6

import spacy
from collections import Counter

#Load medium NLP model.
processor = spacy.load("en_core_web_md")

#Define text to analyze.
text = """'Early detection of Alzheimer's disease (AD) during its long prodromal period is of paramount importance clinically. Understanding how language changes during this period may provide a deeper understanding of early progression of AD and strengthen early detection when treatments may most effectively halt disease progression (Golde, DeKosky, & Galasko, 2018; Selkoe, 2021). At the same time, such investigation can inform our understanding of language as a cognitive system and crucially advance our knowledge of fundamental brain-language relations.Study of the neuroscientific foundations of the progression of AD, including its prodromal preclinical periods, has advanced widely (Braak & Braak, 1991, 1996; Braak, Thal, Ghebremedhim, & Del Tredicit, 2011; Dickerson et al., 2008, 2011, 2013; Dickerson & Wolk, 2012; Karaman, Mormino, & Sabuncu, 2022; Putcha, Brickhouse, Wolk, & Dickerson, 2019; Sperling et al., 2011). Behaviorally and clinically, Mild Cognitive Impairment (MCI) is defined as a prodromal stage in which there is evidence of cognitive decline that often continues to advance to dementia. Primary difficulty with memory during this period, referred to as MCI amnestic subtype (aMCI), is most likely to be attributable to Alzheimer's disease (AD) pathology (Petersen, 2003, 2004; Rye, Vik, Kocinski, Lundervold, & Lundervold, 2022; Vos et al., 2013; Wolk, Dickerson, & Alzheimer’s Disease Neuroimaging Initiative, 2011). However, while memory deficits are primary, there are also decades of research revealing that deficits in other cognitive domains, including language, also occur during prodromal and clinical periods of AD (see Sherman, Henderson, Flynn, Gair, & Lust, 2021 for recent review). Within the research on language, there are differing views as to the degree and manner in which linguistic factors, syntactic, semantic, or both (Ahmed, Haigh, de Jager, & Garrard, 2013; Chapin, Clarke, Garrard, & Hinzen, 2022; Joyal, Groleau, Bouchard, Wilson, & Fecteau, 2020; Kemper et al., 1993) contribute to this decline.'"""

#Process text.
processed_text2 = processor(text)

#Define target word.
relate_to_this = "disease"

#Counter that holds the semantic relatedness scores.
semantic_relatedness = Counter()

#Obtain semantic relatedness score by analyzing each LOWER CASE token in processed_text2.
for token in processed_text2:
    similarity = token.similarity(processor.vocab[relate_to_this])
    semantic_relatedness[token.text.lower()] = similarity

#Get the most semantically related words.
semantically_related_words = semantic_relatedness.most_common(10)

#Print the most semantically related words to the target word.
print("Most semanitcally related word to '{}' in the body of text:".format(relate_to_this))
for word, similarity in semantically_related_words:
    print(f"{word}: {similarity}")
```
